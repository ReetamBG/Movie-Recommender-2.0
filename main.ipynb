```python
# title
df['title'] = df['title'].apply(lambda x : x.replace('#', ''))
```

```python
df = df[['movie_id', 'poster_path', 'title', 'overview', 'keywords', 'genres', 'cast', 'crew']]
df.head()
```

```python
# clearing out missing values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df = df.reset_index(drop=True)    # cuz few columns were dropped 
```

### Creating tags

```python
# overview
df['overview'] = df['overview'].str.replace(',', '').str.replace('.', '')
df['overview'] = df['overview'].str.split()
df['overview']
```

```python
# keywords
df['keywords'].values[0]
```

```python
df['keywords'] = df['keywords'].apply(lambda x : eval(x))                                  # extracting list from string representation
df['keywords'] = df['keywords'].apply(lambda x : [word.replace(' ', '') for word in x])    # removing spaces between words if any
df['keywords']
```

```python
# genres
df['genres'].values[0]
```

```python
def extract_values(li, key):
    values = []
    for i in li:
        values.append(i[key])
    return values
```

```python
df['genres'] = df['genres'].apply(lambda x : eval(x))     # extracting list from string representation
df['genres']
```

```python
df['genres'] = df['genres'].apply(extract_values, key='name')
df['genres']
```

```python
# cast 
df['cast'] = df['cast'].apply(lambda x : eval(x))     # extracting list from string representation
df['cast']
```

```python
df[df['title'] == 'Batman']['cast'].values[0]
```

```python
def extract_top3_values(li, key):
    values = []
    j = 0
    for i in li:
        if j < 3:
            values.append(i[key])
            j += 1
    return values
```

```python
df['actors'] = df['cast'].apply(extract_top3_values, key='name')
df['actors']
```

```python
df['actors'] = df['actors'].apply(lambda x : [word.replace(' ', '') for word in x])    # removing spaces between words if any
df['actors']
```

```python
df['characters'] = df['cast'].apply(extract_top3_values, key='character')
df['characters'] = df['characters'].apply(lambda x : [word.replace(' ', '') for word in x])
df['characters']
```

```python
# crew
df['crew'] = df['crew'].apply(lambda x : eval(x))     # extracting list from string representation
df['crew'].values[0]
```

```python
def extract_director(li):
    values = []
    for i in li:
        if i['job'] == 'Director':
            values.append(i['name'])
    return values
```

```python
df['director'] = df['crew'].apply(extract_director)
df['director'] = df['director'].apply(lambda x : [word.replace(' ', '') for word in x])
df['director']
```

```python
df.head()
```

```python
# Creating the tags

tags = df['overview'] + df['keywords'] + df['genres'] + df['actors'] + df['characters'] + df['director'] 
tags
```

# Vectorization - Bag of Words

### Preprocessing

```python
import nltk
nltk.download('stopwords')
```

```python
# stemming

from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language="english", ignore_stopwords=True)     

import time
start = time.time()
tags_stemmed = tags.apply(lambda x : [stemmer.stem(word) for word in x])
print(time.time() - start)
```

```python
# converting back into string for vectorization

tags_stemmed = tags_stemmed.apply(lambda x : ' '.join(x))
tags_stemmed
```

```python
# vectorization

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000, lowercase=True, stop_words='english', dtype=np.int8)

vectors = vectorizer.fit_transform(tags_stemmed).toarray()
vectors
```

```python
# cosine similarity matrix 

# from sklearn.metrics.pairwise import cosine_similarity

# start = time.time()
# similarity_matrix = cosine_similarity(vectors)
# print(time.time() - start)
```

```python
# cosine similarity matrix calculation in chunks

from sklearn.metrics.pairwise import cosine_similarity
import os
import pickle

def cosine_similarity_chunks(vectors=vectors, chunk_size=300):

    output_dir = 'similarity_matrix'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    num_movies = vectors.shape[0]
    similarity_matrix = np.empty((num_movies, num_movies))

    for i in range(0, num_movies, chunk_size):
        similarity_matrix_chunk = cosine_similarity(vectors[i : i+chunk_size], vectors)
        similarity_matrix[i: i+chunk_size] = similarity_matrix_chunk
        
        chunk_no = i // chunk_size
        pickle.dump(similarity_matrix_chunk, open(f'{output_dir}/chunk{chunk_no}.pkl', 'wb'))
        print('Exporting chunk', chunk_no)

    return similarity_matrix
```

```python
similarity_matrix = cosine_similarity_chunks()
```

```python
similarity_matrix.shape
```

# Main Function

```python
def recommend(movie):
    movie_index = df.loc[df['title'] == movie].index[0]
    similarity_vector = similarity_matrix[movie_index]
    top_5 = sorted(list(enumerate(similarity_vector)), key=(lambda x : x[1]), reverse=True)
    top_5 = top_5[1:6]
    top_5_index = [i[0] for i in top_5]
    top_5_movies = df.loc[top_5_index, 'title']
    
    return top_5_index, top_5_movies
```

```python
recommend('Batman')
```
